**Let's Add new feature named "Region" to features in our Population Dataframe using the states column**

Nigeria for instance has states divided into regions, grouping states into regions migh give some insight on what's happening in regional areas

```python
#Creating those regions and their states, we have:
NC = ['Benue', 'Kogi', 'Kwara', 'Nasarawa', 'Niger', 'Plateau', 'FCT']
SS = ['Akwa', 'Bayelsa', 'Cross River', 'Rivers', 'Delta', 'Edo']
SE = ['Abia', 'Anambra', 'Ebonyi', 'Enugu', 'Imo']
SW = ['Ekiti', 'Lagos', 'Ogun', 'Ondo', 'Osun', 'Oyo']
NE = ['Adamawa', 'Bauchi', 'Borno', 'Gombe', 'Taraba', 'Yobe']
NW = ['Jigawa', 'Kaduna', 'Kano', 'Katsina', 'Kebbi', 'Sokoto', 'Zamfara']
```

```python
#Creating those a function that can map states to different regions, we have:
def region(state):
    NC = ['Benue', 'Kogi', 'Kwara', 'Nasarawa', 'Niger', 'Plateau', 'FCT']
    SS = ['Akwa', 'Bayelsa', 'Cross River', 'Rivers', 'Delta', 'Edo']
    SE = ['Abia', 'Anambra', 'Ebonyi', 'Enugu', 'Imo']
    SW = ['Ekiti', 'Lagos', 'Ogun', 'Ondo', 'Osun', 'Oyo']
    NE = ['Adamawa', 'Bauchi', 'Borno', 'Gombe', 'Taraba', 'Yobe']
    NW = ['Jigawa', 'Kaduna', 'Kano', 'Katsina', 'Kebbi', 'Sokoto', 'Zamfara']
    if state in NC:
        return 'North Central'
    elif state in SS:
        return 'South South'
    elif state in SE:
        return 'South East'
    elif state in SW:
        return 'South West'
    elif state in NE:
        return 'North East'
    else:
        return 'North West'
```

```python
#Applying th efunction using pandas.apply() method, we have:

Population_df['Region'] = Population_df['STATES'].apply(region)

#Lets see whats in the new feature and the entire dataset
Population_df
```

**You will notice that all the states EXCEPT FCT are assigned to NORTH WEST which is the last action under the if statement used in the region function**

This is because, each state is accompained by word **State**, and since "Abia" in SE list is not the same as "Abia State" in the column, we need to split the values in the STATES column so that each state stands alone without repeating the word state.

Lets do this :

```python
#Attempting how to use list method to split values in the State Column
#Using a small list as a sample

states = ['Abia State', 'Oyo State', 'Akwa Ibom State']
tri = []  #A new list to store the answer
for state in states:  #Iterate over the list such that each element is referef to as state
    if len(state.split()) <= 2:   #If the length of state is between 0 and 2 ie. Oyo State has len 2
        tri.append(state.split()[0]) #Split it and pick the first part and store in tri
    else:      #If not
        tri.append(state.split()[0] + ' ' + state.split()[1]) #Split it and pick the first part, add space inbetween and pick the second too
        #eg Akwa Ibom State has length 3, split it and take the first 2 elements but add space between them and store in tri
print('the new state list is ', tri)
```

You will notice that this procedure works so well that all the occurence of State beside the state's name has been removed

> Instead to transforming this to a function, let's try using **LIST COMPREHENSION**


```python
Population_df['STATE'] = [state.split()[0] if len(state.split())<=2 else (state.split()[0] + ' ' + state.split()[1]) for state in (Population_df['STATES'])]

Population_df
```

for instance, gender could be 'Male' or 'Female'. This implies that if gender is encoded as a numbers, anytime a gender is Male, the female equivalence == 0 and anytime the Gender is Female

### Merging Datasets
